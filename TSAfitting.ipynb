```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit

# Load the CSV file into a pandas DataFrame
df = pd.read_csv(r'/Users/yourname/Documents/project/data.csv)

# Extract the data from the x and y columns
x_data = df.iloc[:, x].values
y_data = df.iloc[:, y].values

# Define the sigmoid function
def sigmoid(x, L, x0, k, b):
    y = L / (1 + np.exp(-k*(x-x0))) + b
    return y

# Fit the data to the sigmoid function
# Define the x-axis limits for fitting
x_min = 60
x_max = 80
x_fit = x_data[(x_data >= x_min) & (x_data <= x_max)]
y_fit = y_data[(x_data >= x_min) & (x_data <= x_max)]

# Perform curve fitting
p0 = [max(y_fit), np.median(x_fit), 1, min(y_fit)] # Initial guess for the parameters
params, covariance = curve_fit(sigmoid, x_fit, y_fit, p0=p0)

# Extract the fitted parameters
L, x0, k, b = params

# Output the fitted parameters
print("Fitted Parameters:")
print(params)
#print(params([0])) = {Ldimer})
#print(params([1]) = {x0dimer})
#print(params([2]) = {kdimer})
#print(params([3]) = {bdimer})

# Generate x values for plotting the fitted curve
x_fit_curve = np.linspace(min(x_fit), max(x_fit), 1000)
y_fit_curve = sigmoid(x_fit_curve, *params)

# Find y value corresponding to x0 on the fitted curve
y0dim = sigmoid(x0, L, x0, k, b)
x0dim = x0

#generate sigmoidal fit curve for tetramer peak
# Fit the data to the sigmoid function
# Define the x-axis limits for fitting
x_min = 80
x_max = 95
x_fit = x_data[(x_data >= x_min) & (x_data <= x_max)]
y_fit = y_data[(x_data >= x_min) & (x_data <= x_max)]

# Perform curve fitting
p0 = [max(y_fit), np.median(x_fit), 1, min(y_fit)] # Initial guess for the parameters
paramstet, covariance = curve_fit(sigmoid, x_fit, y_fit, p0=p0)

# Extract the fitted parameters
L, x0, k, b = paramstet

# Output the fitted parameters
print("Fitted Parameters:")
print(paramstet)

# Generate x values for plotting the fitted curve
x_fit_curvetet = np.linspace(min(x_fit), max(x_fit), 1000)
y_fit_curvetet = sigmoid(x_fit_curvetet, *paramstet)


import matplotlib.pyplot as plt

from scipy.stats import sem
print(sem(x_fit_curve))

fig, ax=plt.subplots(figsize=(6.5, 5))

x_dat = x_data[(x_data >= 50) & (x_data <= 95)]
y_dat = y_data[(x_data >= 50) & (x_data <= 95)]

#plot of fit for dimer
plt.scatter(x_dat, y_dat, marker ='o', color='b')
plt.plot(x_fit_curve, y_fit_curve, 'b-', '-')

#plot of fit for tetramer
plt.scatter(x_dat, y_dat, marker ='o', color='b')
plt.plot(x_fit_curvetet, y_fit_curvetet, 'b-', '-')

# Find y value corresponding to x0 on the fitted curve
y0 = sigmoid(x0, L, x0, k, b)

# Plot marker for x0
plt.plot()
#plt.legend(x0, y0, 'r.', label='T$_m$ = 75.5', loc='upper right', fontsize=13, title_fontsize=14)
plt.xlabel('Temperature (Â°C)', fontsize = 16, fontweight="bold")
plt.ylabel('Normalized Fluorescence', fontsize = 18, fontweight="bold")

# Set y-axis limits and tick locations
plt.ylim(0, 1)  # Adjust the y-axis limits as needed
ytick_values = np.arange(0.2, 1.1, 0.2)

plt.yticks(ytick_values, [f'{val:.1f}' for val in ytick_values])  # Set the y-axis tick locations and labels
plt.tick_params(labelsize=14.5, pad=1)
ax.spines["top"].set_linewidth(3)
ax.spines["left"].set_linewidth(3)
ax.spines["right"].set_linewidth(3)
ax.spines["bottom"].set_linewidth(3)

plt.tight_layout()
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"

plt.plot(title='TITLE')
plt.plot(x0dim, y0dim, 'k.', label='T$_m$ dimer  = aÂ°C', markersize=15)
plt.plot(x0, y0, 'k.', label='T$_m$ tetramer  = bÂ°C', markersize=15)

#plt.plot(x_dat, y_dat, label='\nTITLE')
#plt.plot(x0, y0, label='\nT$_m$ = a')

plt.legend(loc='upper left', fontsize=13, title_fontsize=14, title='WT')

#plt.title('TITLE')
plt.xlim(60, 95)  # Set x-axis limits
plt.show()
```
